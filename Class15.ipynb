## **Example: Unit & Integration Testing**
The following example simulates a simple 'Order System' where we check if a single function works (Unit Test) and if two components—a Calculator and an Inventory Manager—work together (Integration Test).

```python
# The Application code:
class InventoryManager:
    def __init__(self):
        self.stock = {'apples': 10, 'bananas': 5}

    def has_stock(self, item, quantity):
        return self.stock.get(item, 0) >= quantity

def calculate_total(price, quantity, tax_rate=0.1):
    '''A simple function to calculate total price.'''
    if quantity < 0:
        raise ValueError('Quantity cannot be negative')
    return (price * quantity) * (1 + tax_rate)
```

```python
# Unit Tests: Testing a single function in isolation:
def test_calculate_total_success():
    # Test a standard successful calculation
    assert calculate_total(10, 2) == 22.0

def test_calculate_total_failure():
    # Testing how the function handles bad input:
    with pytest.raises(ValueError):
        calculate_total(10, -1)
```

```python
# Integration tTsts: Testing how two parts work together:
def test_order_process_integration():
    inventory = InventoryManager()
    item = 'apples'
    quantity = 3
    price_per_unit = 2.0

    # Checking if components interact correctly:
    if inventory.has_stock(item, quantity):
        total = calculate_total(price_per_unit, quantity)
        assert total == 6.6  # (2.0 * 3) + 10% tax
    else:
        pytest.fail('Integration failed: Stock should have been available')
```
